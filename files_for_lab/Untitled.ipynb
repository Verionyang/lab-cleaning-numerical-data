```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
customer_df = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
display(customer_df.shape)
display(customer_df.info())
customer_df.head()
```

```python
col = []
for column in customer_df.columns: 
    col.append(str(column).lower().replace(" ","_"))

customer_df.columns = col

customer_df.head()
```

```python
customer_df['effective_to_date']=pd.to_datetime(customer_df['effective_to_date'])
display(customer_df.info())
customer_df.head(5)
```

```python
def num_cat_split(df):
    df_num = df.select_dtypes(include = np.number)
    df_cat = df.select_dtypes(include = 'object')
    return df_num, df_cat
```

```python
customer_num,customer_cat = num_cat_split(customer_df)
```

```python
customer_cat = customer_cat.drop(columns="customer",axis=1)
customer_cat.head(5)
```

```python
fig,ax = plt.subplots(figsize=(7,6))
sns.heatmap(customer_num.corr(),annot=True)
plt.show()
```

The pair of monthly premiums of the auto and the total claim has the highest correlation. It suggests that the higher the premium cost is, the higher the total claim amount is. A higher premium implies a better insurance package so the drivers can declare more claims and get more money in return.

```python
def con_dis_split(df, thr=36):
    con = []
    dis = []
    for i in range(df.shape[1]):
        if df.iloc[:,i].unique().size >=thr: 
            con.append(i)
        else: 
            dis.append(i)
    return df.iloc[:,con],df.iloc[:, dis]
```

```python
customer_con,customer_dis = con_dis_split(customer_num)

display(customer_con.head(5))
customer_dis.head(5)
```

```python
def barplot_df(df):
    fig,ax = plt.subplots(1,df.shape[1],figsize=(16,16))
    for i in range(0,df.shape[1]):
        unique_values = df.iloc[:,i].unique().size
        if (unique_values <= 6):
            sns.countplot(x=df.iloc[:,i], order=df.iloc[:,i].value_counts().index,ax=ax[i])
        else:
            sns.countplot(y=df.iloc[:,i], order=df.iloc[:,i].value_counts().index,ax=ax[i])
    return fig
```

```python
fig1 = barplot_df(customer_dis)
```

```python
def histplot_df(df):
    fig,ax = plt.subplots(1,df.shape[1],figsize=(16,16))
    for i in range(0,df.shape[1]):
        sns.histplot(x=df.iloc[:,i],ax=ax[i])
    return fig
```

```python
fig2 = histplot_df(customer_con)
```

I will apply power transformation to the continous variables to change the pattern so they are closer to Gaussian.. 

```python
def boxplot_df(df):
    fig,ax = plt.subplots(1,df.shape[1],figsize=(16,16))
    for i in range(0,df.shape[1]):
        sns.boxplot(y=df.iloc[:,i],ax=ax[i])
    return fig
```

```python
fig3 = boxplot_df(customer_con)
```
